### Data Cleaning for the Historical Alarm Tag Dataset from category 1

Tag name: OGREY1_G1_INNTAK_RIST_FALLTAP_HH_AL

Alarm Description: 

The following preprocessing steps were performed:
1) Loaded the raw alarm log data.
2) Selected the relevant columns and removed unit columns.
3) Converted the DateTime column from object to datetime format.
4) Converted values to string format using astype(str).
5) Replaced comma decimal separators with dots using str.replace(',', '.').
6) Checked for missing values.
7) Checked for duplicate records.
8) Exported the cleaned dataset.


```python
import pandas as pd
```

```python
historic_data_Rist= pd.read_csv(r"E:\Uis Cs\Master thesis\final_project\data\dataset2(tag _INNTAK_RIST_FALLTAP_HH_AL)\raw_data\export-7-tag_OGREY1_INNTAK_RIST_FALLTAP_HH_AL.csv",sep=";",dtype=str,engine="python")
historic_data_Rist.shape     #  get the shape of dataset 
```

```python

# Make a working copy
clean_historical_data= historic_data_Rist.copy()
```

```python
# Since there are separate columns for each feature, representing units, we remove those unit columns to keep the dataset clean

clean_historical_data= clean_historical_data.loc[:, ~clean_historical_data.columns.str.contains("Unit", case=False)]
```

```python
clean_historical_data.columns=['DateTime','RIST_MAKS_SP_LAST', 'RIST_AUTO_DIFF_SP_LAST(cm)','RIST_MAKS_TIME_TM_PV', 'RIST_FALLTAP_PV(cm)', 'RIST_MAKS_PADRAG_SP_LAST(%)','RIST_MAKS_PADRAG_SP_TM_PV','RIST_AUTT_DIFF_TM_PV(cm)']
```

```python
clean_historical_data.columns
```

```python
clean_historical_data.head(5)   
```

```python
#Creates a list of the columns you want to clean, exclude DateTime
#astype(str)  >> Converts values into text/string format
#.str.replace(',', '.') >> Changes decimal commas into decimal points.

cols=['RIST_MAKS_SP_LAST', 'RIST_AUTO_DIFF_SP_LAST(cm)','RIST_MAKS_TIME_TM_PV', 'RIST_FALLTAP_PV(cm)', 'RIST_MAKS_PADRAG_SP_LAST(%)','RIST_MAKS_PADRAG_SP_TM_PV','RIST_AUTT_DIFF_TM_PV(cm)']

for col in cols:
    clean_historical_data[col] = clean_historical_data[col].astype(str).str.replace(',', '.', regex=False)
    clean_historical_data[col] = clean_historical_data[col].apply(pd.to_numeric, errors='coerce')    # Converts the cleaned text into numeric values (float64).
```

```python
clean_historical_data['DateTime'] = pd.to_datetime(clean_historical_data['DateTime'], errors='coerce')
```

```python
clean_historical_data.info()
```

```python
# # Check for missing values in each column
clean_historical_data.isna().sum()
```

```python
#Check missing values percentage
(clean_historical_data.isna().mean() * 100)
```

```python
# Remove columns with too many missing values

clean_historical_data = clean_historical_data.drop(columns=['RIST_MAKS_TIME_TM_PV', 'RIST_MAKS_PADRAG_SP_TM_PV'])
```

```python
clean_historical_data.columns
```

```python
# Estimate missing values based on neighboring data points
clean_historical_data = clean_historical_data.interpolate()
```

```python
clean_historical_data.isna().sum()
```

```python
clean_historical_data.isna().sum()        #No missing value 
```

```python
# Check for duplicate rows and columns

duplicate_rows = clean_historical_data.duplicated().sum()
duplicate_collumns = clean_historical_data.T.duplicated().sum()
print("Duplciate rows: ", duplicate_rows)
print("Duplicate columns: ", duplicate_collumns)
```
