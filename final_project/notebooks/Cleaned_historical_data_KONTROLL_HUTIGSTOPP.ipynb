### Data Cleaning for the Historical Alarm Tag Dataset from category 1

Tag name: OGREY1_G1_KONTROLL_HURTIGSTOPP_MEK_AL

Alarm Description: Generator 1 was stopped by a mechanical emergency stop (QSD)

The following preprocessing steps were performed:
1) Loaded the raw alarm log data.
2) Selected the relevant columns and removed unit columns.
3) Converted the DateTime column from object to datetime format.
4) Converted values to string format using astype(str).
5) Replaced comma decimal separators with dots using str.replace(',', '.').
6) Checked for missing values.
7) Checked for duplicate records.
8) Exported the cleaned dataset.


```python
import pandas as pd
```

```python
## get the file_path of the dataset
file_path = r"E:\Uis Cs\Master thesis\code\Clean_HistData\final_Hist_data\raw data\KONTROLL_HURTIGSTOPP_MEK_AL.csv"
```

```python
# diaplay the dataset in a pandas dataframe

historic_data = pd.read_csv(file_path,sep=";",dtype=str,engine="python")
historic_data.shape  # get  the shape of dataset 
```

```python
historic_data.head(5)  # shows the first 5 rows of dataset
```

```python
# # get the more information about the dataset ( feature names, missing value count and data type)

historic_data.info()
```

```python

# Make a working copy
analys_historic_data= historic_data.copy()
```

```python
# Since there are separate columns for each feature, representing units, we remove those unit columns to keep the dataset clean

analys_historic_data= analys_historic_data.loc[:, ~analys_historic_data.columns.str.contains("Unit", case=False)]
```

```python
# after removing the  Unit colums so the number of columns become 6
analys_historic_data.columns
```

  Shortet the name of Columns:
    
    'Value (Cluster1.OGREY1_G1_KONTROLL_REG_P_SP_SP_LAST)'     changed to   'REG_P_LAST(KW)',     
    'Value (Cluster1.OGREY1_G1_KONTROLL_REG_NIVA_SP_SP_LAST)'  changed to   'REG_NIVA_LAST(m)',
    'Value (Cluster1.OGREY1_G1_KONTROLL_REG_NIVA_SP_TM_PV)'    changed to   'REG_NIVA_PV(m)',
    'Value (Cluster1.OGREY1_G1_KONTROLL_REG_P_SP_TM_PV)'       changed to   'REG_P_PV(KW)',
    'Value (Cluster1.OGREY1_G1_KONTROLL_AGC_DB_SP)'            changed to   'AGC_DB_SP',

    Since the unit columns were removed, the unit for each feature is appended to the column name in parentheses.


```python
analys_historic_data.columns=['DateTime','REG_P_LAST(KW)', 'REG_NIVA_LAST(m)','REG_NIVA_PV(m)', 'REG_P_PV(KW)', 'AGC_DB_SP']
analys_historic_data.columns
```

```python

analys_historic_data.head()
```

```python
#Creates a list of the columns you want to clean, exclude DateTime
#astype(str)  >> Converts values into text/string format
#.str.replace(',', '.') >> Changes decimal commas into decimal points.

cols=['REG_P_LAST(KW)','REG_NIVA_LAST(m)','REG_NIVA_PV(m)', 'REG_P_PV(KW)', 'AGC_DB_SP']
for col in cols:
    analys_historic_data[col] = analys_historic_data[col].astype(str).str.replace(',', '.', regex=False)
    analys_historic_data[col] = analys_historic_data[col].apply(pd.to_numeric, errors='coerce')    # Converts the cleaned text into numeric values (float64).
```

```python
# The DateTime column converted from object data type to datetime format

analys_historic_data['DateTime'] = pd.to_datetime(analys_historic_data['DateTime'], errors='coerce')
```

```python
analys_historic_data.info()
```

```python
# Shows the numbe rof missing values for each column
analys_historic_data.isna().sum()
```

```python
# estimates values from nearby timestamps
analys_historic_data = analys_historic_data.interpolate()
```

```python
# Check for duplicate rows and columns

duplicate_rows = analys_historic_data.duplicated().sum()
duplicate_collumns = analys_historic_data.T.duplicated().sum()
print("Duplciate rows: ", duplicate_rows)
print("Duplicate columns: ", duplicate_collumns)
```
